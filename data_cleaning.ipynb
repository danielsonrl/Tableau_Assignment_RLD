```python
# Import your dependecies.
import pandas as pd
import numpy as np
import glob
```

```python
# Make a dataframe for the quarterly and monthly data.
quarterly_data_df = pd.concat([pd.read_csv(csvfile) for csvfile in glob.glob("Raw_Data/City_and_Relationship_Data/*.csv")], ignore_index = True)
monthly_data_df = pd.concat([pd.read_csv(csvfile) for csvfile in glob.glob("Raw_Data/Bike_Data/*.csv")], ignore_index = True)
```
